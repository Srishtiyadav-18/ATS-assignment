```python
import requests
from bs4 import BeautifulSoup

def scrape_jobs():
    url = "https://www.indeed.com/jobs?q=Python+Developer&l="
    headers = {"User-Agent": "Mozilla/5.0"}
    response = requests.get(url, headers=headers)

    if response.status_code == 200:
        soup = BeautifulSoup(response.text, "html.parser")
        jobs = []
        for job in soup.find_all("div", class_="job_seen_beacon"):
            title = job.find("h2").text.strip() if job.find("h2") else "N/A"
            company = job.find("span", class_="companyName").text.strip() if job.find("span", class_="companyName") else "N/A"
            salary = job.find("div", class_="salary-snippet-container")
            salary = salary.text.strip() if salary else "N/A"

            jobs.append({"title": title, "company": company, "salary": salary})

        return jobs
    else:
        print("Failed to retrieve data")
        return []

job_list = scrape_jobs()
print(job_list)
```

```python
from pymongo import MongoClient
import requests
from bs4 import BeautifulSoup


def scrape_jobs():
    url = "https://www.indeed.com/jobs?q=Python+Developer&l="
    headers = {"User-Agent": "Mozilla/5.0"}
    response = requests.get(url, headers=headers)

    if response.status_code == 200:
        soup = BeautifulSoup(response.text, "html.parser")
        jobs = []
        for job in soup.find_all("div", class_="job_seen_beacon"):
            title = job.find("h2").text.strip() if job.find("h2") else "N/A"
            company = job.find("span", class_="companyName").text.strip() if job.find("span", class_="companyName") else "N/A"
            salary = job.find("div", class_="salary-snippet-container")
            salary = salary.text.strip() if salary else "N/A"

            jobs.append({"title": title, "company": company, "salary": salary})

        return jobs
    else:
        print("Failed to retrieve data")
        return []

# Defining job_list within the same scope before calling save_to_mongodb()
job_list = scrape_jobs()
print(job_list)


client = MongoClient("mongodb://localhost:27017/")
db = client["job_database"]
collection = db["jobs"]

def save_to_mongodb(jobs):
    if jobs:
        collection.insert_many(jobs)
        print("Data saved successfully!")
    else:
        print("No jobs to save.")

#job_list is defined before this line and in the same scope
save_to_mongodb(job_list)
```
